## Chunk Text data

```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

loader = TextLoader('/content/Commentary sample.txt', encoding = "utf-8")
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100,separators = ".")
docs = text_splitter.split_documents(documents)
```

```python
len(docs)
```

## Create Embeddings

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.embeddings import SentenceTransformerEmbeddings


embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/LaBSE')
```

## Store in ChromaDB

```python
from langchain.vectorstores import Chroma

db = Chroma.from_documents(docs, embeddings)
retriever = db.as_retriever()
```

```python
query = "who is Joseph?"
docs = retriever.get_relevant_documents(query)

#How many sources will be returned as relevant?
print(docs[0])
```

## RAG

```python
import openai
#from kaggle_secrets import UserSecretsClient
#user_secrets = UserSecretsClient()
openai.api_key = "Put Open ai key"
```

```python
from langchain.chat_models import ChatOpenAI
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0, openai_api_key=openai.api_key)
#from langchain.llms import OpenAI
#llm = OpenAI(model_name="text-davinci-003",temperature=0, openai_api_key=openai.api_key)
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.chains import RetrievalQA

qa_chain = load_qa_chain(llm, chain_type="stuff")
qa = RetrievalQA(combine_documents_chain=qa_chain,
                 retriever= retriever)

query = "Who is Joseph?"
qa.run(query)
```
